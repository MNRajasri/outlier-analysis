## <center> No of outliers </center>

```python
q1=np.percentile(visa_df["no_of_employees"],25)
q3=np.percentile(visa_df["no_of_employees"],75)

lb=q1-1.5*(q3-q1)
ub=q1+1.5*(q3-q1)

con1=visa_df["no_of_employees"]<lb
con2=visa_df["no_of_employees"]>ub

visa_df_outliers=visa_df[con1 | con2]
visa_df_outliers
```

## <center> No of non outliers </center>

```python
q1=np.percentile(visa_df["no_of_employees"],25)
q3=np.percentile(visa_df["no_of_employees"],75)

lb=q1-1.5*(q3-q1)
ub=q1+1.5*(q3-q1)

con1=visa_df["no_of_employees"]>lb
con2=visa_df["no_of_employees"]<ub

visa_df_nonoutliers=visa_df[con1 & con2]
visa_df_nonoutliers
```

## <center> Whole data vs Non-Outliers data </center>

```python
plt.figure(figsize=(10,7))
plt.subplot(2,2,1).hist(visa_df["no_of_employees"])
plt.title("Histogram")
plt.subplot(2,2,2).boxplot(visa_df["no_of_employees"],vert=False)
plt.title("Boxplot")
plt.subplot(2,2,3).hist(visa_df_nonoutliers["no_of_employees"])
plt.title("Histogram")
plt.subplot(2,2,4).boxplot(visa_df_nonoutliers["no_of_employees"],vert=False)
plt.title("Boxplot")
plt.show()
```

## <center> How to deal with outliers </center>

- replacing with median

```python
l=[]

q1=np.percentile(visa_df["no_of_employees"],q=25)
q3=np.percentile(visa_df["no_of_employees"],q=75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr

mediann=round(np.median(visa_df["no_of_employees"]),2)
mediann

for i in visa_df["no_of_employees"].values:
    if i<lb or i>ub:
        l.append(mediann)
    else:
        l.append(i)

df1=visa_df.copy()
df1["no_of_employees"]=l
df1
```

```python
plt.boxplot(df1["no_of_employees"],vert=False)
plt.show()
```

## <center> Whole data vs Data replaced with median </center>

```python
plt.figure(figsize=(10,7))
plt.subplot(2,2,1).hist(visa_df["no_of_employees"])
plt.title("Histogram")
plt.subplot(2,2,2).boxplot(visa_df["no_of_employees"],vert=False)
plt.title("Boxplot")
plt.subplot(2,2,3).hist(df1["no_of_employees"])
plt.title("Histogram")
plt.subplot(2,2,4).boxplot(df1["no_of_employees"],vert=False)
plt.title("Boxplot")
plt.show()
```
